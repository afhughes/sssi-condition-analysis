```python
import arcpy
sortFeat  = r'S:\DataUnit\DptShare\Consultancy\Lodge\Conservation and Policy\DavidHampson\Lost_Decade_PA_analysis\ArcGIS Pro\Protected_Area_Analysis_UK\Protected_Area_Analysis_UK.gdb\temp_rank' 
sortField = 'PERC_SSSI_FAV A; SSSI_Ha D' 
#sortField = 'PERC_SSSI_FAV A'
idField   = 'Rank' 
rec=0

def autoIncrement():
    global rec
    pStart    = 1 
    pInterval = 1 
    if (rec == 0): 
        rec = pStart 
    else: 
        rec += pInterval 
    return rec

#expression = u'{} IS NOT NULL'.format(arcpy.AddFieldDelimiters(sortFeat, "PERC_SSSI_FAV"))
rows = arcpy.UpdateCursor(sortFeat,"","","", sortField)

for row in rows:
    
    row.setValue(idField, autoIncrement())
    rows.updateRow(row)

del row, rows
```
